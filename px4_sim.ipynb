```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = "/home/zpyang/data/"
data_file = data_dir + "px4_sitl_log_" + "2024-08-13-015750.npy"

# "time, pos[3], vel[3], att[4], gyro[3], acce[3]"
data = np.load(data_file)
print(data.shape)
time = (data[:,0] - data[0,0])
x = data[:,1]
y = data[:,2]
z = data[:,3]
vx = data[:,4]
vy = data[:,5]
vz = data[:,6]
qw = data[:,7]
qx = data[:,8]
qy = data[:,9]
qz = data[:,10]
w1 = data[:,11]
w2 = data[:,12]
w3 = data[:,13]
a1 = data[:,14]
a2 = data[:,15]
a3 = data[:,16]
```

```python
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
ax.plot(x, y, z, label="")
ax.set_xlabel('X Position (m)')
ax.set_ylabel('Y Position (m)')
ax.set_zlabel('Z Position (m)')
ax.set_title('SITL quadcopter flight trajectory')
# ax.legend()

fig2 = plt.figure()
ax1 = fig2.add_subplot(311)
ax1.plot(time, a1)
ax2 = fig2.add_subplot(312)
ax2.plot(time, a2)
ax3 = fig2.add_subplot(313)
ax3.plot(time, a3)
```
